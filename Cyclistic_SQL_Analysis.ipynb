# Cyclistic Data Cleaning & Exploration SQL Queries

```sql
-- #1: ride_id consistency check
SELECT LENGTH(ride_id), count(*)
FROM `plated-dryad-448619-k8.Cyclistic.Cyclistic_Combined_Table`
GROUP BY LENGTH(ride_id);

SELECT COUNT (DISTINCT ride_id)
FROM `plated-dryad-448619-k8.Cyclistic.Cyclistic_Combined_Table`;
-- Each ride_id string is 16 digits
-- Total ride IDs: 5,860,568; Unique: 5,860,357; Duplicates: 211
```

```sql
-- #2: Distinct rideable types
SELECT DISTINCT rideable_type
FROM `plated-dryad-448619-k8.Cyclistic.Cyclistic_Combined_Table`;
-- classic_bikes, electric_bikes, electric_scooters
```

```sql
-- #3: Identify rides less than 1 minute or more than 1 day
SELECT *
FROM `plated-dryad-448619-k8.Cyclistic.Cyclistic_Combined_Table`
WHERE TIMESTAMP_DIFF(ended_at, started_at, MINUTE) < 1
   OR TIMESTAMP_DIFF(ended_at, started_at, MINUTE) > 1440;
-- 139,124 rides found
```

```sql
-- #4: Null start station names
SELECT start_station_name, count(*)
FROM `plated-dryad-448619-k8.Cyclistic.Cyclistic_Combined_Table`
GROUP BY start_station_name
ORDER BY start_station_name;

-- Null end station names
SELECT end_station_name, count(*)
FROM `plated-dryad-448619-k8.Cyclistic.Cyclistic_Combined_Table`
GROUP BY end_station_name
ORDER BY end_station_name;

-- Distinct station names vs IDs
SELECT COUNT(DISTINCT(start_station_name)) AS Cyl_startname,
       COUNT(DISTINCT(end_station_name)) AS Cyl_endname,
       COUNT(DISTINCT(start_station_id)) AS Cyl_startid,
       COUNT(DISTINCT(end_station_id)) AS Cyl_endid
FROM `plated-dryad-448619-k8.Cyclistic.Cyclistic_Combined_Table`;
-- Start names: 1,808; End names: 1,815; Start IDs: 1,763; End IDs: 1,768
```

```sql
-- #5: Null station names and IDs
SELECT count(*) as num_of_rides
FROM `plated-dryad-448619-k8.Cyclistic.Cyclistic_Combined_Table`
WHERE start_station_name IS NULL OR
      start_station_id IS NULL OR
      end_station_name IS NULL OR
      end_station_id IS NULL;
-- 1,652,259 null values found
```

```sql
-- #6: Null latitude and longitude
SELECT *
FROM `plated-dryad-448619-k8.Cyclistic.Cyclistic_Combined_Table`
WHERE start_lat IS NULL OR
      start_lng IS NULL OR
      end_lat IS NULL OR
      end_lng IS NULL;

SELECT count(*) AS lat_lng
FROM `plated-dryad-448619-k8.Cyclistic.Cyclistic_Combined_Table`
WHERE start_lat IS NULL OR
      start_lng IS NULL OR
      end_lat IS NULL OR
      end_lng IS NULL;
-- 7,232 null coordinates found
```

```sql
-- #7: Member types
SELECT DISTINCT member_casual
FROM `plated-dryad-448619-k8.Cyclistic.Cyclistic_Combined_Table`;
-- Two member types: Casual and Member
```
